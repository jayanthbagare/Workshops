```python
import math
import numpy as np
from activation_functions import activations as act
global data
```

```python
data = np.arange(-5.0,5.0,0.01)
```

```python
def make_plot(x,y):
    import matplotlib.pyplot as plt
    %matplotlib inline
    plt.plot(x,y)
    plt.grid()
    plt.show()
```

```python
# tanh plot
x,y = act(data).tanh()
make_plot(x,y)
```

```python
#relu
x,y = act(data).relu()
make_plot(x,y)
```

```python
#Sigmoid
x,y = act(data).sigmoid()
make_plot(x,y)
```

```python
#Gaussian
x,y = act(data).gaussian()
make_plot(x,y)
```

```python
#Step
x,y = act(data).step()
make_plot(x,y)
```

```python
#Softplus
x,y = act(data).softplus()
make_plot(x,y)
```

```python
#Leaky RELU
x,y = act(data).leaky_relu()
make_plot(x,y)
```
