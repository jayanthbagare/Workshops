#### So the derivative is defined as 
$$
L = lim_{h->0}  f(a+h) - f(h)/h
$$

If you see the above in terms of 
h = 0.001, which is a small increment added to x
then
$$
y_1 = f(x)\\
y_2 = f(x+h)
$$
Then the slope(rate of increase/decrease) is

$$
\frac {y_2-y_1}{x_2-x_1}
$$
Which would be

$$
\frac {f(x+h) - f(x)}{(x+h) - (x)}
$$
Which would become 

$$
\frac {f(x+h) - f(x)}{h}
$$

```python
#Define Derivative function
def der(x,h):
    y1 = curve(x)
    y2 = curve(x+h)
    return (y2-y1)/h
```

```python
y = curve(x)
```

```python
h = 0.001
dy_dx = der(x,h)
```

#### By Addition rule of differentiation
$$
eq = 3x + 10 \\
f'(eq) = f'(3x) + f'(10) \\
       = 3 + 0 \\
       = 3 \\
$$

#### But what if we have a slightly more complex equation(Multivariable)
#### We apply partial differential equations to the fore
$$
eq = 3a + b \\
f'(eq) = \frac{\partial eq}{\partial a b} \\
       = \frac{\partial eq}{\partial a} + \frac{\partial eq}{\partial b} \\
       = \frac{\partial (3a + b)}{\partial a} + \frac{\partial (3a + b)}{\partial b}\\
       = 3 + 1\\
       = 4
$$

#### Let us walk backward from L; i.e. we take the gradient from L and distribute it into the components

#### By product rule
L = f * d
$$
(fg)' = f'g + fg'\\
f' = \frac{\partial L}{\partial f} = d\\
g' = \frac{\partial L}{\partial d} = f\\
(fg)' = \frac{\partial L}{\partial f} * d + f * \frac{\partial L}{\partial d}\\
(fg)' = d + f
$$


$$
L = f * d\\
\frac{\partial L}{\partial d} = f (how)\\
\frac{(x+h)-(x)}{h}\\
\frac{\partial L}{\partial d} = \frac{(d+h)*f - d*f}{h}\\
              = \frac{fd + fh - fd}{h}\\
              = f
$$
