### For a curve line how does the derivative look

```python
def curve(x):
    return x**2
```

```python
x = np.arange(-5.0,5.0,0.1)
y = curve(x)
```

```python
#slope at x = 0,2,4
x0 = 0
x2 = 2
x4 = 4

h = 0.001

y0 = curve(x0)
y1 = curve(x0 + h)

y2 = curve(x2)
y3 = curve(x2 + h)

y4 = curve(x4)
y5 = curve(x4 + h)
```

```python
dy0 = (y1 - y0)/((x0+h) - x0)
dy2 = (y3 - y2)/((x2+h) - x2)
dy4 = (y5 - y4)/((x4+h) - x4)
```

```python
print(f"Slope for {x0} is {dy0}")
print(f"Slope for {x2} is {dy2}")
print(f"Slope for {x4} is {dy4}")
```

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

#### So the derivative is defined as 
$$
L = lim_{h->0}  f(a+h) - f(h)/h\\~\\
$$

If you see the above in terms of 
h = 0.001, which is a small increment added to x
then
$$
y_1 = f(x)\\~\\
y_2 = f(x+h)\\~\\
$$
Then the slope(rate of increase/decrease) is

$$
\frac {y_2-y_1}{x_2-x_1}\\~\\
$$
Which would be

$$
\frac {f(x+h) - f(x)}{(x+h) - (x)}\\~\\
$$
Which would become 

$$
\frac {f(x+h) - f(x)}{h}\\~\\
$$

```python
#Define Derivative function
def der(x,h):
    y1 = curve(x)
    y2 = curve(x+h)
    return (y2-y1)/h
```

```python
y = curve(x)
```

```python
h = 0.001
dy_dx = der(x,h)
```

```python
dy_dx
```

```python
plt.plot(x,y)
plt.plot(x,dy_dx,'--g')
plt.grid()
plt.show()
```

#### By Addition rule of differentiation
$$
eq = 3x + 10 \\~\\
f'(eq) = f'(3x) + f'(10) \\~\\
       = 3 + 0 \\~\\
       = 3 \\~\\
$$

#### But what if we have a slightly more complex equation(Multivariable)
#### We apply partial differential equations to the fore
$$
eq = 3a + b \\~\\
f'(eq) = \frac{\partial eq}{\partial a b} \\~\\
       = \frac{\partial eq}{\partial a} + \frac{\partial eq}{\partial b} \\~\\
       = \frac{\partial (3a + b)}{\partial a} + \frac{\partial (3a + b)}{\partial b}\\~\\
       = 3 + 1\\~\\
       = 4\\~\\
$$

#### Armed with this knowledge, let us go ahead to see it in terms of limit function

```python
# L's gradient with itself is 1, hence we make L.grad to 1.0
L.grad = 1.0
```

#### Let us walk backward from L; i.e. we take the gradient from L and distribute it into the components

#### By product rule
L = f * d
$$
(fg)' = f'g + fg'\\~\\
f' = \frac{\partial L}{\partial f} = d\\~\\
g' = \frac{\partial L}{\partial d} = f\\~\\
(fg)' = \frac{\partial L}{\partial f} * d + f * \frac{\partial L}{\partial d}\\~\\
(fg)' = d + f\\~\\
$$


$$
L = f * d\\~\\
\frac{\partial L}{\partial d} = f (how)\\~\\
\frac{(x+h)-(x)}{h}\\~\\
\frac{\partial L}{\partial d} = \frac{(d+h)*f - d*f}{h}\\~\\
              = \frac{fd + fh - fd}{h}\\~\\
              = f\\~\\
$$

```python
#Therefore 
'''
f.grad = d.data
d.grad = f.data
'''
f.grad = d.data
d.grad = f.data
```

```python
bu.draw_dot(L)
```

So now we have
d = c + e
If we apply the limit
$$
\frac{dd}{dc} = \frac{(f(x+h) - f(x))}{h}\\~\\
              = \frac{(c+e+h)-(c+e)}{h}\\~\\
              = \frac{c+e+h-c-e}{h}\\~\\
              = \frac{h}{h}\\~\\
              = 1
$$

Similarly we get
$$
\frac{dd}{de} = 1
$$

But how is the sensitivity of c on L; well we can apply the chain rule

$$
\frac{dL}{dc} = \frac{dL}{dd} \frac{dd}{dc}
$$

Similarly for sensitivity of e on L
$$
\frac{dL}{de} = \frac{dL}{dd} \frac{dd}{de}
$$

Substituting the values
$$
\frac{dL}{dc} = 2 * 1 = 2\\~\\
\frac{dL}{de} = 2 * 1 = 2\\~\\
$$

Ok now the last step, we use the chain rule all the way
$$
\frac{dL}{da} = \frac{dL}{dd} \frac{dd}{de} \frac{de}{da} \\~\\
\frac{dL}{db} = \frac{dL}{dd} \frac{dd}{de} \frac{de}{db} \\~\\
$$

As proven for the product rule above, the grad is the other's value
i.e

$$
a.grad = b.data\\
b.grad = a.data\\
$$
