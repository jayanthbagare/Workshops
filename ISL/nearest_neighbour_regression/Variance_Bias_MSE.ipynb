```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x = np.linspace(-1.0,1.0,100)
y_lin = 3*x + 2
y_sq = x*x + 2
y_cub = x*x*x + 2
y_quad = x*x*x*x + 2
```

```python
df = pd.DataFrame({'x':x,'y_lin':y_lin,'y_sq':y_sq,'y_cub':y_cub,'y_quad':y_quad})
df
```

```python
def sigmoid(y):
    y = 1 / (1 + np.exp(-(y)))
    return y
```

```python
y_sigmoid = sigmoid(df['y_cub'])
# y_sigmoid
```

```python
fig,axes = plt.subplots(2,2,figsize=(10,5))
sns.scatterplot(ax=axes[0,0],data=df,x=x,y=y_lin)
sns.relplot(ax=axes[1,0],data=df,x=x,y=y_sigmoid,color="red")
sns.scatterplot(ax=axes[0,1],data=df,x=x,y=y_sq)
sns.scatterplot(ax=axes[1,0],data=df,x=x,y=y_cub)
sns.scatterplot(ax=axes[1,1],data=df,x=x,y=y_quad)
```
